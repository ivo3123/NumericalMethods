Here we will take a look at why interpolating with a polynomial of an inadequate degree may lead to undesired results.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util/')

from lagrange_poly import lagrange_poly
```

```python
start = 2
end = 10
```

```python
x_axis = np.linspace(start, end, 300)
```

```python
def f(x):
    return np.sin(2*x) + np.e ** np.cos(x)
```

```python
plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.legend(['f(x)'])
plt.show()
```

>It would not be easy interpolating the function in this interval.

```python
degree = 7

plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.plot(x_axis, lagrange_poly(np.linspace(start, end, degree + 1), f, x_axis))
plt.legend(['f(x)', f'Lagrange poly of {degree}th degree'])
plt.show()
```

>7th degree is still too low and approximation is quite poor.

```python
degree = 15

plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.plot(x_axis, lagrange_poly(np.linspace(start, end, degree + 1), f, x_axis))
plt.legend(['f(x)', f'Lagrange poly of {degree}th degree'])
plt.show()
```

>15th degree approximates the inner parts of the function quite well but has a noticeable error on the right-most and left-most part

```python
degree = 25

plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.plot(x_axis, lagrange_poly(np.linspace(start, end, degree + 1), f, x_axis))
plt.legend(['f(x)', f'Lagrange poly of {degree}th degree'])
plt.show()
```

>The approximation starts to look more precise but there still is a noticeable error.

```python
degree = 38

plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.plot(x_axis, lagrange_poly(np.linspace(start, end, degree + 1), f, x_axis))
plt.legend(['f(x)', f'Lagrange poly of {degree}th degree'])
plt.show()
```

>Now the approximation is quite good.

```python
degree = 60

plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.plot(x_axis, lagrange_poly(np.linspace(start, end, degree + 1), f, x_axis))
plt.legend(['f(x)', f'Lagrange poly of {degree}th degree'])
plt.show()
```

>Oscillations are starting to appear. That is due to the nature of polynomials of heigher degree.

```python
degree = 64

plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.plot(x_axis, lagrange_poly(np.linspace(start, end, degree + 1), f, x_axis))
plt.legend(['f(x)', f'Lagrange poly of {degree}th degree'])
plt.show()
```

>Now it is quite apperant that the higher the degree does not equal better approximation.

```python
degree = 106

plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.plot(x_axis, lagrange_poly(np.linspace(start, end, degree + 1), f, x_axis))
plt.legend(['f(x)', f'Lagrange poly of {degree}th degree'])
plt.show()
```

>The oscillations now cause exponentially large errors.

## Conclusion

It is quite important to work with algebraic polynomials of an adequate degree.

If the degree is too low, the interpolated polynomial may not describe the behaviour of the function well.

Howerever if the degree is too heigh, the interpolated polynomial may have misleading values due to the oscillations.
