Да се приближи функцията $ f(x) = cos(x) $:
* като се построи интерполационен полином на Лагранж с възли $ 0, \space \pi/4, \space \pi/3 $
* като се построи интерполационен полином на Ермит с възли 0 и $ \pi/3 $ и съответни кратности 2 и 1

Да се построят графиките на относителната грешка (по абсолютна стойност) в интервала $ [0, \pi/3] $.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util')

from hermit_poly import  hermit_poly
from lagrange_poly import lagrange_poly
```

```python
def f(x):
    return np.cos(x)
```

```python
lagrange_nodes = np.array([0, np.pi/4, np.pi/3])

hermit_nodes = np.array([0, 0, np.pi/3])
hermit_values = np.array([1, 0, 0.5])
```

```python
x_axis = np.linspace(0, np.pi/3, 350)

plt.plot(x_axis, f(x_axis), color='blue', linestyle='dashed')
plt.plot(x_axis, lagrange_poly(lagrange_nodes, f, x_axis), color='red')
plt.plot(x_axis, hermit_poly(hermit_nodes, hermit_values, x_axis), color='green')
plt.legend(['f(x) = cos(x)', 'Lagrange poly', 'Hermit poly'])
plt.show()
```

```python
def get_relative_error(function, interpolating_poly, nodes, values, x):
    return abs(function(x) - interpolating_poly(nodes, values, x)) / abs(f(x))
```

```python
plt.plot(x_axis, get_relative_error(f, lagrange_poly, lagrange_nodes, f, x_axis))
plt.plot(x_axis, get_relative_error(f, hermit_poly, hermit_nodes, hermit_values, x_axis))
plt.legend(['Lagrange poly relative error', 'Hermit poly relative error'])
plt.show()
```
