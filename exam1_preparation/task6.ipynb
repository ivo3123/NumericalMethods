Define the function *lagrange_poly*(n, x0, h, f, x), which constructs the Lagrange interpolation polyomial of $n^{th}$ degree for the function $\ f(x) $ with {$\ x_i = x_0 + ih $,   $\ i = 0, 1, ..., n $ }, using the Lagrange formula.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util/')

from lagrange_poly import lagrange_poly
```

```python
def fake_lagrange_poly(degree, x0, h, function, x):
    return lagrange_poly(np.array([x0 + i * h for i in range(0, degree + 1)]), function, x)
```

```python
def f(x):
    return np.cos(x)
```

```python
nodes = np.array([0.3, 0.5, 0.7, 0.9, 1.1, 1.3, 1.5, 1.7, 1.9])

x_axis = np.linspace(nodes.min(), nodes.max(), 200)

plt.scatter([nodes], [f(nodes)])
plt.plot(x_axis, f(x_axis), linestyle='dashed')
plt.plot(x_axis, lagrange_poly(nodes, f, x_axis))
plt.plot(x_axis, fake_lagrange_poly(degree := nodes.size - 1, x0 := 0.3, h := 0.2, f, x_axis))
plt.legend(['points', 'f(x) = cos(x)', 'Lagrange poly', 'Fake Lagrange poly'])
plt.show()
```

As expected, the new Lagrange method and the original Lagrange method produce the same result.
