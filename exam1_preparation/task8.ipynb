The given data describes the absorption of a medicine by the human body.

The table shows the concentration of the medicine in the blood as a function of time.

Find a generalized polynomial of an adequate basis that describes the process.

|Time|Concentration|
|:--:|:-----------:|
| 0  | 0.1         |
| 2  | 0.009       |
| 4  | 0.0011      |
| 6  | 0.00003     |

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util/')

from basis_poly import basis_poly, trig_poly, alg_poly, exp_poly
```

```python
nodes = np.array([0, 2, 4, 6, 8])
values = np.array([0.1, 0.009, 0.0011, 0.00003, 0.0000012])

x_axis = np.linspace(0, 8, 300)
```

```python
plt.scatter(nodes, values)
plt.plot(x_axis, basis_poly(nodes, values, x_axis, lambda x, n: 1 / (1 + n + x)))
plt.title('Basis is 1 / (n + 1 + x)')
plt.legend(['data', 'poly of the basis'])
plt.show()
```

> Very reasonable approximation.

```python
plt.scatter(nodes, values)
plt.plot(x_axis, trig_poly(nodes, values, x_axis))
plt.title('Basis is the trigonometric basis')
plt.legend(['data', 'trig poly'])
plt.show()
```

> The behaviour of this function does not describe the input data well.

```python
plt.scatter(nodes, values)
plt.plot(x_axis, alg_poly(nodes, values, x_axis))
plt.title('Basis is x ^ n')
plt.legend(['data', 'alg poly'])
plt.show()
```

> Somewhat reasonable approximation, however the first polynomial describes the data better.

```python
plt.scatter(nodes, values)
plt.plot(x_axis, basis_poly(nodes, values, x_axis, lambda x, n: np.e ** (-n * x)))
plt.title('Basis is e ^ (-n * x)')
plt.legend(['data', 'poly of the basis'])
plt.show()
```

> Not the polynomial we are looking for.

```python
plt.scatter(nodes, values)
plt.plot(x_axis, exp_poly(nodes, values, x_axis))
plt.title('Basis is e ^ (n * x)')
plt.legend(['data', 'exp poly'])
plt.show()
```

> Not the polynomial we are looking for.

The basis { $\ 1/(1 + x) $, $\ 1/(2 + x) $, ..., $\ 1/(n + x) $ } seems to describe the behaviour of the data **the best**.
