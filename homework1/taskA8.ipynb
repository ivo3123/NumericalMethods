Approximate the data from the table above using the least squares method with a function of the kind:
$ y = c \dfrac{x}{d + x} $

Linearize the model using an adequate transformation.

Visualize.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util')

from least_squares_method import least_squares_method_function
```

```python
nodes = np.array([0.75, 2, 3, 4, 6, 8, 8.5])
values = np.array([1.2, 1.95, 2, 2.4, 2.4, 2.7, 2.6])
```

```python
def f(arr, x):
    return arr[0] * (x / (arr[1] + x))
```

```python
x_axis = np.linspace(nodes.min(), nodes.max(), 200)

plt.scatter(nodes, values, color='orange')
plt.plot(x_axis, least_squares_method_function(nodes, values, x_axis, 2, desired_kind_function=f), color='blue')
plt.title('Approximating data using least squares method')
plt.legend(['points', '$y = \\dfrac{c x}{d+x}$'])
plt.show()
```
