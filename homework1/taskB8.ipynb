Using the least squares method approximate the data from the table with a function of the kind:
$ y = a10 ^{bx} $

Use an adequate linearization of the model.
Visualize graphically.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util')

from least_squares_method import least_squares_method_function
```

```python
nodes = np.array([0.4, 0.8, 1.2, 1.6, 2, 2.3])
values = np.array([800, 975, 1500, 1960, 2900, 3600])
```

```python
def f(arr, x):
    return arr[0] * 10 ** (arr[1] * x)
```

```python
x_axis = np.linspace(nodes.min(), nodes.max(), 200)

plt.scatter(nodes, values, color='orange')
plt.plot(x_axis, least_squares_method_function(nodes, values, x_axis, 2, desired_kind_function=f), color='blue')
plt.title('Approximating data using least squares method')
plt.legend(['points', '$y = a 10^{b x}$'])
plt.show()
```
