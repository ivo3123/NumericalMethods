Find approximately
$ \int_0^{\pi} f(x)dx $,
using Simpson's rule, when given only the following values ​​of the function:

| x | f(x) |
| :-: | :-: |
| 0 | 1.0000 |
| $\dfrac{\pi}{4}$ | 0.3431 |
| $\dfrac{\pi}{2}$ | 0.2500 |
| $\dfrac{3\pi}{4}$ | 0.3431 |
| $\pi$ | 1.0000 |

```python
import numpy as np
```

```python
nodes = np.array([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi])
values = np.array([1, 0.3431, 0.25, 0.3431, 1])
```

```python
a = nodes.min()
b = nodes.max()

n = nodes.size // 2

integral_value = ((b - a) / (6 * n)) * sum(values[i-2] + 4*values[i-1] + values[i] for i in range(2, nodes.size, 2))

integral_value
```
