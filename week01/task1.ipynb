Constuct the graphs of the functions in one coordinate sistem
$$ f(x)=e^x, \space \space \space g(x)=1+x+0.5x^2+0.1667x^3 $$
in the interval $[-0.5,0.5]$.
Construct the graphs of the absolute error and relative error, given by the approximation of $f(x)$ with $g(x)$.

Note that $g(x)$ is a subsequence of the Taylor series for f(x).

$
e^x = \displaystyle\sum_{n=0}^{\infty} \dfrac{x^n}{n!}
$

$
e^x = \dfrac{x^0}{0!} + \dfrac{x^1}{1!} + \dfrac{x^2}{2!} + \dfrac{x^3}{3!} + O(x^4)
$

$
e^x = 1 + x + \dfrac{x^2}{2} + \dfrac{x^3}{6} + O(x^4)
$

$
e^x \approx 1 + x + 0.5 x^2 + 0.1667 x^3
$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.e ** x

g = lambda x: 1 + x + 0.5 * x ** 2 + 0.1667 * x ** 3

start = -0.5
end = 0.5
```

```python
x_axis = np.linspace(start, end, 200)
```

```python
plt.plot(x_axis, f(x_axis), color='purple', linestyle='dashed')
plt.plot(x_axis, g(x_axis), color='lime')
plt.legend(['f(x)', 'g(x)'])
plt.show()
```

```python
absolute_error = lambda f, g, x: abs(f(x) - g(x))

relative_error = lambda f, g, x: abs(f(x) - g(x)) / abs(f(x))
```

```python
plt.plot(x_axis, absolute_error(f, g, x_axis))
plt.title('Absolute error of g(x) approximating f(x)')
plt.show()
```

```python
plt.plot(x_axis, relative_error(f, g, x_axis))
plt.title('Relative error of g(x) approximating f(x)')
plt.show()
```
