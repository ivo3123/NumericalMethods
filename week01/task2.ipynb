The algebraic polynomial is given:
$$p(x)=(x-2)^9=x^9-18x^8+144x^7-672x^6+2016x^5-4032x^4+5376x^3-4608x^2+2304x-512.$$

Construct the graph of $p(x)$ by representing the polynomial the following way:

1) $ \space p(x)=(x-2)^9 $

2) $ \space p(x)=x^9-18x^8+144x^7-672x^6+2016x^5-4032x^4+5376x^3-4608x^2+2304x-512 $.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p1 = lambda x: (x - 2) ** 9

p2 = lambda x: (
    x**9 - 18 * x**8 + 144 * x**7 - 672 * x**6 + 2016 * x**5 +
    -4032 * x**4 + 5376 * x**3 - 4608 * x**2 + 2304 * x - 512
)
```

```python
x_axis = np.linspace(1.9, 2.1, 1000)
```

```python
plt.plot(x_axis, p1(x_axis))
plt.show()

plt.plot(x_axis, p2(x_axis))
plt.show()
```

> The second graph looks more inacurrate. That is because $ p_2 $ needs many more aritmethical operations to be performed which leads to more floating point errors.
