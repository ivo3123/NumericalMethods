The table presents data for the population of the USA in the period 1920-1990.

Construct an interpolation polynomial of seventh degree for the table.

Approximate the population for the following year and compare them to the actual values.

|year|population|
|:-:|:-:|
|1952|157|
|1974|214|
|2000|281.42|

The data:

|year|population|
|:-:|:-:|
|1920|106.46|
|1930|123.08|
|1940|132.12|
|1950|157.27|
|1960|180.67|
|1970|205.05|
|1980|227.23|
|1990|249.46|

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util')

from newton_poly import newton_poly
```

```python
nodes = np.array([1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990])
values = np.array([106.46, 123.08, 132.12, 152.27, 180.67, 205.05, 227.23, 249.46])
```

```python
x_axis = np.linspace(nodes.min() - 15, nodes.max() + 10, 300)

plt.scatter(nodes, values, color='blue')
plt.scatter([1952, 1974, 2000], [157, 214, 281.42], color='red')
plt.plot(x_axis, newton_poly(nodes, values, x_axis), linestyle='dashed', color='black')
plt.legend(['initial data', 'data to check', 'interpolation poly'])
plt.show()
```

The polynomial seems to describe the behaiour of the data well in the range $[1920, 1990]$.

However the polynomial definitely does not describe the data well ouside that interval.

```python
print("1952:", "estimated value:", newton_poly(nodes, values, 1952), 'real value:', 157)
print("1974:", "estimated value:", newton_poly(nodes, values, 1974), 'real value:', 214)
print("2000:", "estimated value:", newton_poly(nodes, values, 2000), 'real value:', 281.42)
```
