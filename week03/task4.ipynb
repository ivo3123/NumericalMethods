The function of Runge is given.

$ f(x) = \dfrac{1}{1 + 25x^2} $

Approximate $ f(x) $ in the inteval $ x \in [-1, 1] $, using an interpolation polynomial of $4^{\text{th}}$ and $10^{\text{th}}$ degree with equidistant nodes. Construct their graphs and the graphs of the absolute error (by absolute value) of the approximation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util')

from newton_poly import newton_poly
```

```python
runge = lambda x: 1 / (1 + 25 * (x ** 2))
```

```python
nodes_degree_10 = np.linspace(-1, 1, 11)
nodes_degree_4 = np.linspace(-1, 1, 5)

values_degree_10 = np.array(runge(nodes_degree_10))
values_degree_4 = np.array(runge(nodes_degree_4))
```

```python
x_axis = np.linspace(-1, 1, 1000)

plt.plot(x_axis, runge(x_axis), label='runge', linestyle='dashed', color='orange')
plt.plot(x_axis, newton_poly(nodes_degree_10, values_degree_10, x_axis), label='deg_10', color='blue')
plt.plot(x_axis, newton_poly(nodes_degree_4, values_degree_4, x_axis), label='deg_4', color='green')
plt.legend(['Runge function', '$10^{th}$ degree interpolation poly', '$4^{th}$ degree interpolation poly'])
plt.show()
```

```python
absolute_error = lambda x, nodes, values: abs(runge(x) - newton_poly(nodes, values, x))
```

```python
plt.plot(x_axis, absolute_error(x_axis, nodes_degree_10, values_degree_10))
plt.plot(x_axis, absolute_error(x_axis, nodes_degree_4, values_degree_4))
plt.legend(['Absolute error of $10^{th}$ degree poly', 'Absolute error of $4^{th}$ degree poly'])
plt.show()
```
