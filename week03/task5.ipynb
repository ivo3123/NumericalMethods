It is well known that the best nodes for interpolation in the interval $ [-1, 1] $ are the so called Tschebyscheff nodes.

The function of Runge is given.

$ f(x) = \dfrac{1}{1 + 25x^2} $

Approximate $ f(x) $ in the inteval $ x \in [-1, 1] $, using an interpolation polynomial $10^{\text{th}}$ degree with the Tshcebyscheff nodes. Construct its graph and the graph of the absolute error (by absolute value) of the approximation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util')

from newton_poly import newton_poly
from tschebyscheff import get_tschebyscheff_nodes
```

```python
runge = lambda x: 1 / (1 + 25 * (x ** 2))
```

```python
nodes = get_tschebyscheff_nodes(11)
values = runge(nodes)
```

```python
x_axis = np.linspace(-1, 1, 1000)

plt.plot(x_axis, runge(x_axis), label='runge', linestyle='dashed', color='orange')
plt.plot(x_axis, newton_poly(nodes, values, x_axis), label='deg_10', color='blue')
plt.legend(['Runge function', '$10^{th}$ degree interpolation poly with Tschebyscheff nodes'])
plt.show()
```

```python
absolute_error = lambda x: abs(runge(x) - newton_poly(nodes, values, x))
```

```python
plt.plot(x_axis, absolute_error(x_axis))
plt.legend(['Absolute error of $10^{th}$ degree poly with Tschebyscheff nodes'])
plt.show()
```
