Determine the value of $\cos(52^o)$, by constructing the Lagrange interpolation polynomial for the function $\cos(x)$ with nodes $45^o,\space 50^o, \space 55^o, \space 60^o$ using Newton's forward interpolation formula.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util')

from newton_forward_poly import newton_forward_poly
```

```python
f = lambda x: np.cos(x)
```

```python
nodes = np.linspace(45*np.pi/180, 60*np.pi/180, 4)
values = f(nodes)

target_x = 52*np.pi/180
```

```python
x_axis = np.linspace(nodes.min(), nodes.max(), 300)

plt.scatter(nodes, values, color='purple')
plt.scatter([target_x], [f(target_x)], color='green')
plt.scatter([target_x], [newton_forward_poly(45*np.pi/180, 5*np.pi/180, 4, np.cos, target_x)], color='red')
plt.plot(x_axis, f(x_axis), linestyle='dashed', color='black')
plt.plot(x_axis, newton_forward_poly(45*np.pi/180, 5*np.pi/180, 4, np.cos, x_axis))
plt.legend(['nodes', f'exact value of f({target_x})', f'approximated value of f({target_x})', 'f(x)', 'interpolation poly'])
plt.show()
```

```python
real_value = f(target_x)
approx_value = newton_forward_poly(45*np.pi/180, 5*np.pi/180, 4, np.cos, target_x)
error = abs(real_value - approx_value)
print(f'f({target_x}) is', real_value)
print(f'Newton_forward_3(f; {target_x}) is', approx_value)
print('Absolute error (by absolute value) is', error)
```
