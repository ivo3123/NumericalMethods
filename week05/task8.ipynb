Construct a interpolation polynomial of an adequate basis for the given data from the table.

Possible basis include:

{$1, e^x, e^{2x}, ..., e^{nx}$}

{$\dfrac{1}{1 + x}, \dfrac{1}{2 + x}, ..., \dfrac{1}{n + x}$}

{$1, e^{-x}, e^{-2x}, ..., e^{-nx}$}

{$\dfrac{1}{1 - x}, \dfrac{1}{2 - x}, ..., \dfrac{1}{n - x}$}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../util')

from basis_poly import basis_poly
```

```python
nodes = np.array([0, 2, 4, 6, 8])
values = np.array([0.1, 0.009, 0.0011, 0.00003, 0.0000012])
```

```python
basis1 = lambda x, n: np.e ** (n * x)
basis2 = lambda x, n: 1 / (1 + n + x)
basis3 = lambda x, n: np.e ** (-n * x)
basis4 = lambda x, n: 1 / (1 + n - x)
```

```python
x_axis = np.linspace(nodes.min(), nodes.max(), 100)
```

```python
plt.scatter(nodes, values)
plt.plot(x_axis, basis_poly(nodes, values, x_axis, basis1))
plt.show()
```

```python
plt.scatter(nodes, values)
plt.plot(x_axis, basis_poly(nodes, values, x_axis, basis2))
plt.show()
```

```python
plt.scatter(nodes, values)
plt.plot(x_axis, basis_poly(nodes, values, x_axis, basis3))
plt.show()
```

It is clear that the first and the third basis are not adequate.

The fourth basis allows for zero in the denominator and thus is not suitable.

The second basis describes the behaviour of the data well.
