Extract the data from the CO_2_data.csv file.

Construct a linear function to approximate the data using the least squares method.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import sys
sys.path.append('../util')

from least_squares_method import least_squares_method_function
from least_squares_method_matrix_form import least_squares_method_matrix_poly
```

```python
file = pd.read_csv('../data/CO_2_data.csv')
```

```python
nodes = file.loc[:, 'year']
values = file.loc[:, 'co2']
```

```python
x_axis = np.linspace(nodes.min(), nodes.max(), 300)

plt.scatter(nodes, values, color='pink')
plt.plot(x_axis, least_squares_method_function(nodes, values, x_axis, 2), linestyle='dashed', color='purple')
plt.plot(x_axis, least_squares_method_matrix_poly(nodes, values, x_axis, 1), linestyle='dotted', color='green')
plt.title('Linear function to approximate the data')
plt.legend(['data', 'linear function', 'linear function'])
plt.show()
```
