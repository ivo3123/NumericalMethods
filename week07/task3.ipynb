Extract the data from the wind_turbine_data_sample.csv file.

Construct a function of third degree to approximate the data using the least squares method.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import sys
sys.path.append('../util')

from least_squares_method import least_squares_method_function
from least_squares_method_matrix_form import least_squares_method_matrix_poly
```

```python
file = pd.read_csv('../data/wind_turbine_data_sample.csv')
```

```python
nodes = file.loc[:, 'wind speed']
values = file.loc[:, 'power output']
```

```python
x_axis = np.linspace(nodes.min(), nodes.max(), 600)

plt.scatter(nodes, values, color='pink')
plt.plot(x_axis, least_squares_method_function(nodes, values, x_axis, 4), linestyle='dashed', color='purple')
plt.plot(x_axis, least_squares_method_matrix_poly(nodes, values, x_axis, 3), linestyle='dotted', color='green')
plt.title('A function of third degree to approximate the data')
plt.legend(['data', 'function of 3rd degree', 'function of 3rd degree'])
plt.show()
```
